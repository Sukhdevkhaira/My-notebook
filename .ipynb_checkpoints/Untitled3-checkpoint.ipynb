## Cost functions and gradient descent

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow.compat.v1 as tf

tf.disable_eager_execution()
number_of_points = 200

x_point = []
y_point = []

a = 0.22
b = 0.78

for i in range(number_of_points):
    x = np.random.normal(0.0, 0.5)
    y = a * x + b + np.random.normal(0.0, 0.1)

    x_point.append([x])
    y_point.append([y])

x_point = np.array(x_point)
y_point = np.array(y_point)
plt.plot(x_point, y_point, 'o', label='Input Data')
plt.legend()
plt.show()
A = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
B = tf.Variable(tf.zeros([1]))
y = A * x_point + B
cost_function = tf.reduce_mean(tf.square(y - y_point))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(cost_function)
model = tf.global_variables_initializer()
with tf.Session() as session:
    session.run(model)

    for step in range(0, 21):
        session.run(train)

        if step % 5 == 0:
            plt.plot(
                x_point,
                y_point,
                'o',
                label='step={}'.format(step)
            )

            plt.plot(
                x_point,
                session.run(A) * x_point + session.run(B)
            )

            plt.legend()
            plt.show()
```

```python
import numpy as np
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

from tensorflow.keras.datasets import mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()
x_train = x_train.reshape(-1, 784).astype(np.float32) / 255.0
x_test = x_test.reshape(-1, 784).astype(np.float32) / 255.0
y_train = np.eye(10)[y_train]
y_test = np.eye(10)[y_test]
train_pixels = x_train[:100]
train_list_values = y_train[:100]
test_pixels = x_test[:10]
test_list_of_values = y_test[:10]
train_pixel_tensor = tf.placeholder(tf.float32, [None, 784])
test_pixel_tensor = tf.placeholder(tf.float32, [784])
distance = tf.reduce_sum(
    tf.abs(train_pixel_tensor - test_pixel_tensor),
    axis=1
)
pred = tf.argmin(distance, 0)
accuracy = 0.0
with tf.Session() as sess:
    for i in range(len(test_list_of_values)):
        nn_index = sess.run(
            pred,
            feed_dict={
                train_pixel_tensor: train_pixels,
                test_pixel_tensor: test_pixels[i]
            }
        )
        print(
            "Test No:", i,
            "Predicted Class:", np.argmax(train_list_values[nn_index]),
            "True Class:", np.argmax(test_list_of_values[i])
        )
        if np.argmax(train_list_values[nn_index]) == np.argmax(test_list_of_values[i]):
            accuracy += 1.0 / len(test_pixels)

    print("Result =", accuracy)
```

## Data clustering

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd

# Parameters
num_vectors = 1000
num_clusters = 4

# Generate training data
x_values = []
y_values = []

for i in range(num_vectors):
    if np.random.random() > 0.5:
        x_values.append(np.random.normal(0.4, 0.7))
        y_values.append(np.random.normal(0.2, 0.8))
    else:
        x_values.append(np.random.normal(0.6, 0.4))
        y_values.append(np.random.normal(0.8, 0.5))

vector_values = np.array(list(zip(x_values, y_values)))

# Plot input data
plt.scatter(x_values, y_values)
plt.title("Input Data")
plt.show()

# K-Means
kmeans = tf.compat.v1.estimator.experimental.KMeans(
    num_clusters=num_clusters,
    use_mini_batch=False
)

def input_fn():
    return tf.data.Dataset.from_tensor_slices(vector_values.astype(np.float32)).batch(100)

# Train
num_steps = 100
for _ in range(num_steps):
    kmeans.train(input_fn)

# Get cluster assignments
cluster_indices = list(kmeans.predict_cluster_index(input_fn))

# Plot clustered data
colors = ['r', 'g', 'b', 'y']

for i in range(num_vectors):
    plt.scatter(
        vector_values[i][0],
        vector_values[i][1],
        c=colors[cluster_indices[i] % len(colors)]
    )

plt.title("K-Means Clustering Result")
plt.show()
```

## Building the training Set

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import pandas as pd

# Parameters
num_vectors = 1000
num_clusters = 4
num_steps = 100

# Initialize lists
x_values = []
y_values = []
vector_values = []

# Generate random data
for i in range(num_vectors):
    if np.random.random() > 0.5:
        x_values.append(np.random.normal(0.4, 0.7))
        y_values.append(np.random.normal(0.2, 0.8))
    else:
        x_values.append(np.random.normal(0.6, 0.4))
        y_values.append(np.random.normal(0.8, 0.5))

# Combine x and y values
vector_values = np.array(list(zip(x_values, y_values)), dtype=np.float32)

# Convert to TensorFlow tensor
vectors = tf.constant(vector_values)

print("Shape:", vectors.shape)
print(vectors[:5])

# Plot data
plt.figure(figsize=(6, 6))
plt.scatter(x_values, y_values, color="blue", label="Input Data")
plt.title("Training Set")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

## Cost functions and optimization

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd

# Parameters
num_vectors = 2000
num_clusters = 4

# Generate random data
x_values = []
y_values = []

for i in range(num_vectors):
    if np.random.random() > 0.5:
        x_values.append(np.random.normal(0.4, 0.7))
        y_values.append(np.random.normal(0.2, 0.8))
    else:
        x_values.append(np.random.normal(0.6, 0.4))
        y_values.append(np.random.normal(0.8, 0.5))

# Create dataset
vector_values = np.column_stack((x_values, y_values))

# Train K-Means
kmeans = KMeans(
    n_clusters=num_clusters,
    random_state=42,
    n_init=10
)

assignment_values = kmeans.fit_predict(vector_values)

# Colors
colors = ["red", "blue", "green", "yellow"]
labels = [colors[i] for i in assignment_values]

# DataFrame
df = pd.DataFrame({
    "x": x_values,
    "y": y_values,
    "color": labels
})

# Plot
plt.figure(figsize=(7,7))
plt.scatter(df["x"], df["y"], c=df["color"], s=20)

# Centroids
centroids = kmeans.cluster_centers_
plt.scatter(
    centroids[:,0],
    centroids[:,1],
    marker="X",
    s=250,
    c="black",
    label="Centroids"
)

plt.title("K-Means Clustering")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
